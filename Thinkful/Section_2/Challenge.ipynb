## Challenge - What Do You See?

This challenge uses a United Nations Oil and Gas Commodities dataset.

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

### Calling .csv to pandas dataframe, checking column names, and renaming.

```python
df = pd.read_csv('UNdata_Export_20180131_OilGas.csv')
df.columns
df.rename(columns={'Country or Area':'Country_or_Area',
                  'Comm. Code': 'Commodity_Code',
                  'Trade (USD)': 'Trade_USD',
                  'Weight (kg)': 'Weight_kg',
                  'Quantity Name': 'Quantity_Name'}, inplace=True)
```

### Previewing data

```python
df.head()
```

```python
df.describe()
```

```python
df.Commodity.value_counts().sort_values().plot(kind = "bar")
plt.ylabel('Values')
plt.title('Commodity');
```

```python
df.groupby('Commodity').Trade_USD.mean().sort_values().plot(kind= "bar");
plt.ylabel('USD in Billions')
plt.title('Commodity');
```

```python
df.groupby('Country_or_Area').Trade_USD.mean().sort_values().plot(kind= "bar");
plt.ylabel('USD in 10s of Billions')
plt.title('Country');
```
