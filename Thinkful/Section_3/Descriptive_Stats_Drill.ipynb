```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

 #### Bernoulli vs. Binomial
Bernoulli Distribution works well to illustrate the likelihood of a binary outcome one event a time.

Binary appears to work like Bernoulli but is more cumulative. Binary takes into consideration the 'n' number of times the trial is ran.

Both apper to describe centralized tendencies. While the Bernoulli set is empty in the middle it still describes a central tendency

```python
plt.subplot(1,2,1)
bernoulli= np.random.binomial(1, .5, 100)
plt.hist(bernoulli)
plt.title('Bernoulli')
plt.axvline(bernoulli.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.mean() + bernoulli.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean() - bernoulli.std(), color='orange', linestyle='dashed', linewidth=2)

plt.subplot(1,2,2)
binomial = np.random.binomial(20, 1/3, 100)
plt.hist(binomial)
plt.title('Binomial')
plt.axvline(binomial.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='orange', linestyle='dashed', linewidth=2)
plt.tight_layout()
```

#### Poisson

The descriptive statistics appear to skew to the left in the Poisson model.

```python
p = np.random.poisson(2, 100)
plt.hist(p)
plt.axvline(poisson.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='orange', linestyle='dashed', linewidth=2)
plt.show()
```

#### Gamma Distribution
used to show likelihood of event occuring over time.

The Gamma distribution skews slightly to the left yet correctly clusters around the larger portion of data.

```python
gamma = np.random.gamma(5,1, 100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='orange', linestyle='dashed', linewidth=2)
plt.show()
```

#### Chi Square

With Chi Square distribution there appears to a higher central tendency the larger the number of samples.

```python
plt.subplot(1,3,1)
chi_sq = np.random.chisquare(2,100)
plt.hist(chi_sq)
plt.axvline(chi_sq.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(chi_sq.mean() + chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(chi_sq.mean() - chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)

plt.subplot(1,3,2)
chi_sq = np.random.chisquare(50,100)
plt.hist(chi_sq)
plt.axvline(chi_sq.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(chi_sq.mean() + chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(chi_sq.mean() - chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)

plt.subplot(1,3,3)
chi_sq = np.random.chisquare(75,100)
plt.hist(chi_sq)
plt.axvline(chi_sq.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(chi_sq.mean() + chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(chi_sq.mean() - chi_sq.std(), color='orange', linestyle='dashed', linewidth=2)
plt.tight_layout()
```

#### Pareto
Pareto has a tendency to skew to the left as it is designed to do so.

```python
plt.subplot(1,3,1)
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 100) + 1) * m
#Display the histogram of the samples, along with the probability density function:

count, bins, _ = plt.hist(s, 100, normed=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='orange')
plt.axvline(s.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(s.mean() + s.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(s.mean() - s.std(), color='orange', linestyle='dashed', linewidth=2)
plt.tight_layout()

plt.subplot(1,3,3)
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 100) + 1) * m
#Display the histogram of the samples, along with the probability density function:

count, bins, _ = plt.hist(s, 100, normed=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='orange')
plt.axvline(s.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(s.mean() + s.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(s.mean() - s.std(), color='orange', linestyle='dashed', linewidth=2)
plt.tight_layout()
```

```python
plt.subplot(1,3,1)
test1 = np.random.binomial(5, .5, 100)
plt.hist(test1)
plt.title('test1')

plt.subplot(1,3,2)
test2 = np.random.binomial(10, 1, 100)
plt.hist(test2)
plt.title('test2');

plt.subplot(1,3,3)
#test2 = np.random.binomial(10, 1, 100)
test3 = test1+test2
plt.hist(test3)
plt.title('test1 + test2');
plt.axvline(test3.mean(), color='orange', linestyle='solid', linewidth=2)
plt.axvline(test3.mean() + test3.std(), color='orange', linestyle='dashed', linewidth=2)
plt.axvline(test3.mean() - test3.std(), color='orange', linestyle='dashed', linewidth=2)
plt.tight_layout()
```
