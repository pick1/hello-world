```python
pop1_3 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
plt.subplot(1,2,1)
plt.hist(pop1_3, alpha=0.5, label='Population 1_3')
plt.axvline(pop1_3.mean(), color='cyan', linestyle='solid', linewidth=2)
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.axvline(pop2.mean(), color='orange', linestyle='solid', linewidth=2)
plt.legend(loc='upper right')

sample1_3 = np.random.choice(pop1_3, 1000, replace=True)
plt.subplot(1,2,2)
plt.hist(sample1_3, alpha=0.5, label='sample 1 @ .3')
plt.axvline(sample1_3.mean(), color='cyan', linestyle='solid', linewidth=2)
plt.axvline(sample1_3.mean() + sample1_3.std(), color='cyan', linestyle='dashed', linewidth=2)
plt.axvline(sample1_3.mean() - sample1_3.std(), color='cyan', linestyle='dashed', linewidth=2)
plt.legend(loc='upper right')

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1_3, equal_var=False))

plt.tight_layout()
```

I changed my test to a chi square distribution and it does not appear to describe in as much visual detail the closeness of the p-value as the binomial distribution test. The chi square does show whether a realtionship exists but it's difficult to see the details.
